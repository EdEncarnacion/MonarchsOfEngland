```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
wessex = ['Egbert ', 'Aethelwulf', 'Aethelbald', 'Aethelbert', 'Aethelred I', 'Alfred the Great', 'Edward the Elder', 'Athelstan', 'Edmund', 'Edred', 'Edwy', 'Edgar', 'Edward the Martyr', 'Ethelred II the Unready', 'Edmund lronside', 'Edward the Confessor']
jelling = ['Cnut (Canute)', 'Harold I Harefoot', 'Harthacanut']
godwin = ['Harold II']
normandy = ['William I', 'William II', 'Henry I']
blois = ['Stephen']
plantagenet_algevin = ['Henry II', 'Richard I', 'John', 'Henry II', 'Edward I', 'Edward II', 'Edward III', 'Richard II']
lancaster = ['Henry IV', 'Henry V', 'Henry VI']
york = ['Edward IV', 'Edward V', 'Richard III']
tudor = ['Henry VII', 'Henry VIII', 'Edward VI', 'Mary I', 'Elizabeth I']
stuart = ['James I', 'Charles I', 'Charles II', 'James II', 'Mary I', 'Anne']
orange_nassau = ['William III']
hanover = ['George I', 'George II', 'George III', 'George IV', 'William IV', 'Victoria']
saxe_coburg_gotha = ['Edward VII']
windsor = ['George V', 'Edward VIII', 'George VI', 'Elizabeth II']

wessexdf = pd.DataFrame(wessex, columns=['name'])
wessexdf['Royal House'] = 'Wessex'

jellingdf = pd.DataFrame(jelling, columns = ['name'])
jellingdf['Royal House'] = 'Jelling'

godwindf = pd.DataFrame(godwin, columns = ['name'])
godwindf['Royal House'] = 'Godwin'

normandydf = pd.DataFrame(normandy, columns = ['name'])
normandydf['Royal House'] = 'Normandy'

bloisdf = pd.DataFrame(blois, columns = ['name'])
bloisdf['Royal House'] = 'Blois'

plantagenet_algevindf = pd.DataFrame(plantagenet_algevin, columns = ['name'])
plantagenet_algevindf['Royal House'] = 'Plantagenet-Algevin'

lancasterdf = pd.DataFrame(lancaster, columns = ['name'])
lancasterdf['Royal House'] = 'Lancaster'

yorkdf = pd.DataFrame(york, columns = ['name'])
yorkdf['Royal House'] = 'York'

tudordf = pd.DataFrame(tudor, columns = ['name'])
tudordf['Royal House'] = 'Tudor'

stuartdf = pd.DataFrame(stuart, columns = ['name'])
stuartdf['Royal House'] = 'Stuart'

orange_nassaudf = pd.DataFrame(orange_nassau, columns = ['name'])
orange_nassaudf['Royal House'] = 'Orange-Nassau'

hanoverdf = pd.DataFrame(hanover, columns = ['name'])
hanoverdf['Royal House'] = 'Hanover'

saxe_coburg_gothadf = pd.DataFrame(saxe_coburg_gotha, columns = ['name'])
saxe_coburg_gothadf['Royal House'] = 'Saxe-Coburg and Gotha'

windsordf = pd.DataFrame(windsor, columns = ['name'])
windsordf['Royal House'] = 'Windsor'
```

```python
royal_houses = pd.concat([wessexdf, jellingdf, godwindf, normandydf, bloisdf, plantagenet_algevindf, lancasterdf, yorkdf, tudordf, stuartdf, orange_nassaudf, hanoverdf, saxe_coburg_gothadf, windsordf])
royal_houses.reset_index(drop=True)
```

```python
plt.rcParams['figure.figsize'] = (40,7) 
sns.set_theme(style='whitegrid')
sns.set_context('poster')
sns.countplot(x='Royal House', data = royal_houses)
plt.xticks(rotation =45)
plt.show()
```
